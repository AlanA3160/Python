```python
#IMPORT STATEMENTS

#FUNCTION DEFINITIONS
#Def add_expense(expense_list)
def add_expense(expense_list):
    #Ask user for the Category, Vendor, Date
    category = input("Enter the category:").lower()
    vendor = input("Enter the vendor:").lower()
    date = input("Enter the date:")
    #Use try/except to get a dollar amount
    try:
        amount = float(input("Enter Amount: "))
    except ValueError:
        print("invalid amount, no record saved")
        return
    #Create a dictionary for the single expense
    expense = {"category": category,
               "vendor": vendor,
               "date": date,
               "amount": amount
               }
    #Append dictionary to the main expense_list
    expense_list.append(expense)
    #Print appended saved
    print("Expense Saved")

#Def show_report(expense_list, filter_type):
def show_report(expense_list, filter_type):
#Set total to 0.0
    total = 0.0
    search_term = ""
#For/IF Statements to Iterate through the list(All, Category, Vendor,)
    if filter_type == "category":
#input for Category
     search_term = input("Enter Category").lower()
#input for Vendor
    elif filter_type == "vendor":
     search_term = input("Enter Vendor").lower()
#Sum Amounts based on filter_type (All, Category, Vendor)
    print(f"\n---{filter_type.upper()} REPORT ---")
    for expense in expense_list:
        if filter_type == "all" or expense[filter_type] == search_term:
#Print formatted total (2 decimal places)
            print(f"{expense['category']}\t{expense['vendor']}\t{expense['date']}\t${expense['amount']:.2f}")
            total += expense['amount']
    print("-" * 25)
    print(f"The total for {search_term if search_term else 'All'} is: ${total:.2f}")


#PROGRAM LOGIC
# Initialize list (all_expenses)
all_expenses = []
# While true loop for continuous entry
while True:
  print("1. Add Expense")
  print("2. Show Report")
  print("3. Quit")
# Print Add, Total, Category, vendor, Quit
# user input for choice
  choice = input("Select an option: ")
# If statement for handling user input, call functions
  if choice == "1":
    add_expense(all_expenses)
  elif choice == "2":
    print("\n---Totaling Options---:")
    print("A. All Expenses")
    print("C. Expenses by Category")
    print("V. Expenses by Vendor")
    filter_type = input("Enter your choice: ").upper()

    if filter_type == "A":
     show_report(all_expenses, "all")
    elif filter_type == "C":
     show_report(all_expenses, "category")
    elif filter_type == "V":
     show_report(all_expenses, "vendor")
    else:
     print("Invalid choice. Please try again.")
  elif choice == "3":
    print("Goodbye!")
    break

```
